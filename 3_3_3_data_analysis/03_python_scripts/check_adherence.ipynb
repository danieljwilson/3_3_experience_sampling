## `HealthKitSamples`

```python
# create empty column
df = df.assign(HealthKitSamples = 0)
df = df.assign(ToHanoi = 0)
df = df.assign(Trailmaking = 0)
df = df.assign(Location = 0)
df = df.assign(HRVar = 0)
df = df.assign(SKdeviceUsage = 0)
df = df.assign(SKkeyboardMetrics = 0)
df = df.assign(SKmessagesUsage = 0)
df = df.assign(SKspeechTelephony = 0)
df = df.assign(SKvisits = 0)
```

```python
df.head(3)
```

```python
check_df = week4
```

```python
# load datafile 
# change to loop through all files

for i in range(len(check_df)):
    # HealthKit
    hk_df = pd.read_csv(path + check_df[i] + '/HealthKitSamples_' + check_df[i][-17:] + '.csv')
    df['HealthKitSamples'][df.ParticipantIdentifier.isin(np.unique(hk_df.ParticipantIdentifier))] = 1
    # Tower of Hanoi
    hk_toh = pd.read_csv(path + check_df[i] + '/SurveyTowerOfHanoiResults_' + check_df[i][-17:] + '.csv')
    df['ToHanoi'][df.ParticipantIdentifier.isin(np.unique(hk_toh.ParticipantIdentifier))] = 1
    # Trailmaking
    hk_tm = pd.read_csv(path + check_df[i] + '/SurveyTrailmakingResults_' + check_df[i][-17:] + '.csv')
    df['Trailmaking'][df.ParticipantIdentifier.isin(np.unique(hk_tm.ParticipantIdentifier))] = 1
    # Location Visits
    hk_loc = pd.read_csv(path + check_df[i] + '/AppleLocationVisits_' + check_df[i][-17:] + '.csv')
    df['Location'][df.ParticipantIdentifier.isin(np.unique(hk_loc.ParticipantIdentifier))] = 1
    # HR Var
    hk_hrVar = pd.read_csv(path + check_df[i] + '/HealthKitSamples_' + check_df[i][-17:] + '.csv')
    hk_hrVar = hk_hrVar.loc[hk_hrVar.Type == 'HeartRateVariability']
    hk_hrVar = hk_hrVar.drop_duplicates(subset=['ParticipantIdentifier'])
    df['HRVar'][df.ParticipantIdentifier.isin(np.unique(hk_hrVar.ParticipantIdentifier))] = 1
    # SK device usage
    sk_path = path + check_df[i] + '/' + 'sensorkit-device-usage/iPHone'
    sk_participants = os.listdir(os.chdir(sk_path))
    df['SKdeviceUsage'][df.ParticipantIdentifier.isin(sk_participants)] = 1
    # SK keyboard metrics
    sk_path = path + check_df[i] + '/' + 'sensorkit-keyboard-metrics/iPhone'
    sk_participants = os.listdir(os.chdir(sk_path))
    df['SKkeyboardMetrics'][df.ParticipantIdentifier.isin(sk_participants)] = 1
    # SK message usage
    sk_path = path + check_df[i] + '/' + 'sensorkit-messages-usage/iPhone'
    sk_participants = os.listdir(os.chdir(sk_path))
    df['SKmessagesUsage'][df.ParticipantIdentifier.isin(sk_participants)] = 1
    # SK telephony
    sk_path = path + check_df[i] + '/' + 'sensorkit-speech-telephony/iPhone'
    sk_participants = os.listdir(os.chdir(sk_path))
    df['SKspeechTelephony'][df.ParticipantIdentifier.isin(sk_participants)] = 1
    # SK telephony
    sk_path = path + check_df[i] + '/' + 'sensorkit-visits/iPhone'
    sk_participants = os.listdir(os.chdir(sk_path))
    df['SKvisits'][df.ParticipantIdentifier.isin(sk_participants)] = 1
```
